```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
bank_marketing = fetch_ucirepo(id=222)

# data (as pandas dataframes)
X = bank_marketing.data.features
y = bank_marketing.data.targets

# metadata
print(bank_marketing.metadata)

# variable information
print(bank_marketing.variables)
```

Now, let's examine the distributions of the categorical features.

```python
# Select categorical columns from X
categorical_cols = X.select_dtypes(include=['object']).columns

# Plot count plots for categorical features
# Adjusting figsize for better readability, especially with many categories
fig, axes = plt.subplots(nrows=len(categorical_cols), ncols=1, figsize=(12, 6 * len(categorical_cols)))
fig.suptitle('Distribution of Categorical Features', y=1.02)

for i, col in enumerate(categorical_cols):
    if len(categorical_cols) > 1:
        ax = axes[i]
    else:
        ax = axes
    sns.countplot(data=X, y=col, order=X[col].value_counts().index, palette='viridis', ax=ax)
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel('Count')
    ax.set_ylabel(col)

plt.tight_layout()
plt.show()
```
